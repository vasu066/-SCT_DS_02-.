```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/tested.csv")
df
```

# Task
Analyze the relationships between variables in df, identify patterns and trends in the data, and provide the code for this analysis.

## Correlation analysis

### Subtask:
Calculate and visualize the correlation matrix of the numerical variables in `df`.


**Reasoning**:
Select the numerical columns, calculate the correlation matrix, and visualize it using a heatmap.



```python
numerical_df = df.select_dtypes(include=np.number)
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

## Categorical variable relationships

### Subtask:
Analyze the relationships between categorical variables in `df` and the target variable (Survived) using cross-tabulations and visualizations.


**Reasoning**:
Create cross-tabulations between categorical variables and 'Survived', then visualize them using bar plots.



```python
sex_survived_crosstab = pd.crosstab(df['Sex'], df['Survived'])
pclass_survived_crosstab = pd.crosstab(df['Pclass'], df['Survived'])
embarked_survived_crosstab = pd.crosstab(df['Embarked'], df['Survived'])

sex_survived_crosstab.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Survival Count by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()

pclass_survived_crosstab.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Survival Count by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()

embarked_survived_crosstab.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Survival Count by Embarked')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

## Numerical and categorical variable relationships

### Subtask:
Explore the relationships between numerical and categorical variables in `df` using visualizations such as box plots or violin plots.


**Reasoning**:
Create box plots to visualize the relationship between 'Survived' and 'Age', and 'Pclass' and 'Fare'.



```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```
